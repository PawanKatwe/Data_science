```python
df2 = df.drop(columns=['Unnamed: 0'])
```

```python
df2.sort_values(by = ['User.ID'],ascending=False)
```

```python
df2 = df2.rename(columns = {'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating' })
```

```python
#number of unique users in the dataset

len(df2.User_ID.unique())

```

```python
#number of unique Books in the dataset

len(df2.Book_Title.unique())
```

```python
userid_books = pd.pivot_table(df2,index = "User_ID", columns = "Book_Title", values = "Book_Rating").reset_index(drop=True)
```

```python
userid_books
```

```python
userid_books.fillna(0,inplace=True)
```

```python
userid_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_similarity = 1 - pairwise_distances(userid_books.values,metric='cosine')
```

```python
user_similarity
```

```python
#Store the results in a dataframe
user_similarity_df = pd.DataFrame(user_similarity)
```

```python
user_similarity_df.describe()
```

```python
#Set the index and column names to user ids 
user_similarity_df.index = df2.User_ID.unique()
user_similarity_df.columns = df2.User_ID.unique()
```

```python
user_similarity_df
```

```python
import numpy as np


np.fill_diagonal(user_similarity, 0)
user_similarity_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
sim_df = pd.DataFrame(user_similarity_df.idxmax(axis=1))
```
