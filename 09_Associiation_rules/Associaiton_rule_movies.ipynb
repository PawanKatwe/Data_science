```python
## Taking columns with movies name as it represents the same data which column 1 to 5 has
## In absense of infomation assuming we are assuming 
    # Each row indicates the one person 
    # 0 : Movie not watched
    # 1 : watched the movie 
```

```python
df2 = df.iloc[:,5:]
```

```python
import seaborn as sns
```

```python
frequent_itemsets = apriori(df2, min_support = 0.4 ,use_colnames = True)
frequent_itemsets

#support : support of an item % (or number) of transaction in which antecedent(if)
#          and consequet (then) appears in the data 

## list of moveis with support 40% or more
```

```python
rules = association_rules(frequent_itemsets, metric='lift',min_threshold=0.7)

rules.sort_values('lift' , ascending=False)
```

```python
rules_df = rules
```

```python
unique_rules=rules_df.iloc[::2].sort_values('lift', ascending=False).reset_index().drop(columns="index")
unique_rules


#top unique rules 
```
