```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df = pd.read_csv(r'F:\pk\01_Data_science\ASSIGNMENTS\09_Association_rule\groceries.csv',
            header=None,sep=";")

#we have all the shop items in one transactino in single row
```

```python
import numpy as np
```

```python
#taking each rows as list with single item list

list_values=df.values.tolist()
```

```python
#each row as item in list
list2=[]
for sublist in list_values:
    for item in sublist:
        list2.append(item)
        
#each row as list with grocery items
list3=[]
for i in list2:
    list3.append(i.split(","))
    
#list consisting of all grocery items 
final_list=[]
for sublist in list3:
    for item in sublist:
        final_list.append(item)
        
#list consisting of only uniqe grocery items
unique_item=[]
for x in final_list:
    if x not in unique_item:
        unique_item.append(x)
```

```python
# creating a data frame with unique items as columns 

df2 = pd.DataFrame(columns=unique_item)

df3=df2.copy()
```

```python
#Assigning 0 or 1 

for index, row in df.iterrows():
    items = str(row[0]).split(',')
    one_hot_encoding = np.zeros(len(unique_item),dtype=int)
    for item_name in items:
        for i,column in enumerate(df3.columns):
            if item_name == column:
                one_hot_encoding[i] = 1
    df3.at[index] = one_hot_encoding

# Transction encoder is fastest method to do all this.
```

```python
frequent_itemsets= apriori(df3, min_support=0.03, use_colnames=True)
frequent_itemsets
```

```python
rule = association_rules(frequent_itemsets, metric="lift", min_threshold=0.3)

rule_df = rule.iloc[::2].sort_values("lift", ascending=False).reset_index().drop(columns="index")
```
